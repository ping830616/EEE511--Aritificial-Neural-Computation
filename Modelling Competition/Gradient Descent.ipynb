```python
from numpy import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import operator

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from scipy.interpolate import *

data = pd.read_csv('IPG2211A2N.csv',index_col=0)


# second column: 'energy production'
data.columns = ['Energy Production']


data.insert(0, '#', range(1, 1 + len(data)))
data
```

```python
# the first model
def gradient_descent(x,y):
    
    a_curr = 0 
    b_curr = 0
    c_curr = 0
   
    iterations = 1
    n = len(x)
    learning_rate = 0.05
    
    for i in range(iterations):
        y_predicted = a_curr*(x**2) + b_curr*x + c_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])      # MSE
        ad = -(2/n)*sum((x**2)*(y-y_predicted))
        bd = -(2/n)*sum(x*(y-y_predicted))
        cd = -(2/n)*sum(y-y_predicted)
        
        a_curr = a_curr - learning_rate * ad
        b_curr = b_curr - learning_rate * bd
        c_curr = c_curr - learning_rate * cd   
        print("a {}, b {}, c {}, cost {} iteration {}".format(a_curr,b_curr,c_curr,cost, i))
        
        
        
x = np.array(data['#'])
y = np.array(data['Energy Production'])
    
gradient_descent(x,y)
```
